```python
#12-df.groupby(): Agrupa os dados em um DataFrame por uma ou mais colunas.


df = pd.read_excel('fato.xlsx')

df_agrupado = df.groupby("Produto")['Quantidade'].sum().reset_index()

dfframe = pd.DataFrame(df_agrupado)

display(dfframe)
```

```python
#13-df.merge(): Mescla dois DataFrames com base em uma ou mais chaves.
fato = pd.read_excel("fato.xlsx")
dim = pd.read_excel("dimensao.xlsx")

df_mesclado = fato.merge(dim, on ="ID Cliente")

df_mesclado= pd.DataFrame(df_mesclado)

display(df_mesclado)
```

```python
#14-df.concat(): Concatena dois ou mais DataFrames ao longo de um eixo

df_concatenado = pd.concat([fato,dim])
df_concatenado = pd.DataFrame(df_concatenado)
display(df_concatenado)
```

```python
#15-df.pivot_table(): Cria uma tabela dinâmica.

df["Data"] = pd.to_datetime(df["Data"])
df["Nome do mes"] = df["Data"].dt.month

pivot = df.pivot_table(index="Produto",columns="Nome do mes",values="Valor Total",aggfunc="sum")

display(pivot)
```

```python
#21- df.rename(): Renomeia colunas ou índices no DataFrame.

df_renomeado = dim.rename(columns={"Sexo":"Genero"})

display(df_renomeado)
```

```python
#22--df.apply(): Aplica uma função ao longo de um eixo do DataFrame

df_renomeado['Genero_ex'] = df_renomeado["Genero"].apply(lambda x: "Masculino" if pd.notna and x == "M" else "Feminino")

display(df_renomeado)
```

```python
#24--df.map(): Mapeia valores de uma Série de acordo com um dicionário ou função.

import pandas as pd

# Exemplo simplificado do DataFrame
data = {
    'ID': [1, 2, 3, 4, 5],
    'sexo': ['M', 'F', 'M', 'F', 'M']
}

df = pd.DataFrame(data)

# Dicionário de mapeamento
map_sexo = {'M': 'Masculino', 'F': 'Feminino'}

# Aplicar o mapeamento usando map()
df['genero_completo'] = df['sexo'].map(map_sexo)

# Mostrar o DataFrame resultante
print(df)

```

```python
#24--df.astype(): Converte o tipo de dados de uma coluna.

import pandas as pd

df = pd.read_excel("fato.xlsx")

df.info()

df = pd.DataFrame(df)

df["Valor Total"] = df["Valor Total"].astype(float)

df.info()

display(df)
```

```python
# 25--Info agrupamento personalizado.

import pandas as pd
df= pd.read_excel("fato.xlsx")
df = pd.DataFrame(df)


Soma_valor = df.groupby("Produto")["Valor Total"].sum().reset_index().sort_values(by="Valor Total",ascending=False)
Soma_valor['Valor Total'] = Soma_valor['Valor Total'].map('{:,.2f}'.format)
conta_valor= df.groupby("Produto")["Quantidade"].sum().reset_index().sort_values(by="Quantidade",ascending=False)
media_venda= df.groupby("Produto")["Valor Total"].mean().reset_index().sort_values(by="Valor Total",ascending=False)
media_venda['Valor Total'] = media_venda['Valor Total'].map('{:,.2f}'.format)
media_qtd =  df.groupby("Produto")["Quantidade"].mean().reset_index().sort_values(by="Quantidade",ascending=False)

print(f'Valor total das transações:\n{Soma_valor}')
print(f'Total de produtos vendidos:\n{conta_valor}')
print(f'Média de produtos vendidos:\n{media_venda}')
print(f'Média de quantidade vendida:\n{media_qtd}')

```

```python
#agrupamento por analise temporal personalisada

df = pd.read_excel("fato.xlsx")
df= pd.DataFrame(df)
#1º é feita a conversão inteira da coluna
df["Data"] = pd.to_datetime(df["Data"])

#2- depois começa a trabalhar com a categoria do tempo 

df["nome_mes"] = df["Data"].dt.month_name()
df["nome_ano"] = df["Data"].dt.year

venda_mes = df.groupby("nome_mes")["Valor Total"].sum().reset_index()
venda_ano = df.groupby("nome_ano")["Valor Total"].sum().reset_index()


print(f'total de vendas para o mês:\n{venda_mes}')
print(f'total de vendas para o ano:\n{venda_ano}')
```

```python
import pandas as pd

# Carregar dados do arquivo ou DataFrame
df = pd.read_excel("fato.xlsx")

# Identificar colunas de valor numérico para correlação
selecao_colunas = df.select_dtypes(include=["int64", "float64"]).columns

# Selecionar colunas específicas para hipóteses de correlação
colunas_hipoteses = ['Valor Total', 'Quantidade']

# Calcular a correlação entre todas as colunas numéricas
correlacao = df[selecao_colunas].corr()

# Calcular a correlação específica para as colunas de hipóteses
correlacao_hipotese = df[colunas_hipoteses].corr()

# Exibir os resultados
print("Correlação entre todas as colunas numéricas:")
print(correlacao)

print("\nCorrelação entre as colunas de hipóteses:")
print(correlacao_hipotese)

print("\nColunas de hipóteses:")
print(colunas_hipoteses)
```
